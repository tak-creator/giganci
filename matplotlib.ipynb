```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
```

```python
x = dziedzina=[-3,-2,-1,0,1,2,3,4]
y = wynik=[(2*x)-6 for x in dziedzina]
plt.plot(x,y, c='r', marker='x', linestyle='--', linewidth=2, markersize=12)
plt.show()
```

```python
# Tworzymy dziedzinę: 400 punktów równomiernie rozmieszczonych między 3 a 100
x = np.linspace(0, 25, 1000)

# Przykładowa funkcja: f(x) = sin(x)
y = np.sin(x)

# Rysowanie wykresu
plt.plot(x, y, label='f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji w przedziale od 0 do 25')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
```

```python
oceny = np.random.randint(1, 6, 31)
plt.hist(oceny, bins=np.arange(0.5, 6.5, 1), alpha=0.7, label='Oceny uczniów')
plt.xlabel('Ocena')
plt.ylabel('Częstotliwość')
plt.title('Rozkład ocen w grupie uczniów')
plt.legend()
plt.show()
```

```python
# Nazwy firm
firmy = ['A', 'B', 'C', 'D', 'E', 'F', 'H']

# Losowe procenty (suma musi dać 100%)
procenty = np.random.rand(7)
procenty = procenty / np.sum(procenty) * 100

# Tworzenie wykresu kołowego
plt.pie(procenty, labels=firmy, autopct='%1.1f%%', startangle=90)
plt.title('Udział firm w rynku')
plt.axis('equal')  # Ustawienie proporcji osi, aby koło było okrągłe
plt.show()
```

```python
# Dane do wykresu
odleglosc = np.arange(0, 5.1, 1)  # Odległość od 0 do 5 z krokiem 0.1
wysokosc = np.random.rand(len(odleglosc)) * 10  # Losowe wysokości od 0 do 10

# Tworzenie wykresu
plt.plot(odleglosc, wysokosc, marker='o', linestyle='-')
plt.xlabel('Odległość')
plt.ylabel('Wysokość')
plt.title('Wykres nierównej drogi')
plt.grid(True)
plt.show()
```
